# DoG

```python
import cv2
import numpy as np
import os
from pathlib import Path

input_folder = r"F:\image_processing\project\dataset\7711810\EuroSAT_RGB\EuroSAT_RGB\AnnualCrop"
output_folder = r"F:\image_processing\project\filtered\Dog\AnnualCrop_DoG"

sigma1 = 1.0
sigma2 = 2.0

os.makedirs(output_folder, exist_ok=True)

for img_path in Path(input_folder).glob("*.jpg"):
    img = cv2.imread(str(img_path), cv2.IMREAD_COLOR)
    if img is None:
        print(f"Skipping {img_path.name}")
        continue

    blur1 = cv2.GaussianBlur(img.astype(np.float32), (0, 0), sigma1)
    blur2 = cv2.GaussianBlur(img.astype(np.float32), (0, 0), sigma2)
    dog = blur1 - blur2

    dog_norm = cv2.normalize(dog, None, 0, 255, cv2.NORM_MINMAX)

    out_path = Path(output_folder) / (img_path.stem + "_dog.png")
    cv2.imwrite(str(out_path), dog_norm.astype(np.uint8))
    print(f"Done: {img_path.name}")

print("All done.")
```

# Sobel

```python
import cv2
import numpy as np
import os
from pathlib import Path

input_folder = r"F:\image_processing\project\dataset\7711810\EuroSAT_RGB\EuroSAT_RGB\AnnualCrop"
output_folder = r"F:\image_processing\project\filtered\Sobel\AnnualCrop_sobel"

os.makedirs(output_folder, exist_ok=True)

for img_path in Path(input_folder).glob("*.jpg"):
    img = cv2.imread(str(img_path), cv2.IMREAD_COLOR).astype(np.float32)
    if img is None:
        print(f"Skipping {img_path.name}")
        continue

    sobel_x = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)
    sobel = np.sqrt(sobel_x**2 + sobel_y**2)

    sobel_norm = cv2.normalize(sobel, None, 0, 255, cv2.NORM_MINMAX)

    out_path = Path(output_folder) / (img_path.stem + "_sobel.png")
    cv2.imwrite(str(out_path), sobel_norm.astype(np.uint8))
    print(f"Done: {img_path.name}")

print("All done.")
```

# Log Transform

```python
import cv2
import numpy as np
import os
from pathlib import Path

input_folder = r"F:\image_processing\project\dataset\7711810\EuroSAT_RGB\EuroSAT_RGB\AnnualCrop"
output_folder = r"F:\image_processing\project\filtered\LT\AnnualCrop_LT"

os.makedirs(output_folder, exist_ok=True)

for img_path in Path(input_folder).glob("*.jpg"):
    img = cv2.imread(str(img_path), cv2.IMREAD_COLOR).astype(np.float32)
    if img is None:
        print(f"Skipping {img_path.name}")
        continue

    log = np.log1p(img)

    log_norm = cv2.normalize(log, None, 0, 255, cv2.NORM_MINMAX)

    out_path = Path(output_folder) / (img_path.stem + "_log.png")
    cv2.imwrite(str(out_path), log_norm.astype(np.uint8))
    print(f"Done: {img_path.name}")

print("All done.")
```

# Negative

```python
import cv2
import os
from pathlib import Path

input_folder = r"F:\image_processing\project\dataset\7711810\EuroSAT_RGB\EuroSAT_RGB\PermanentCrop"
output_folder = r"F:\image_processing\project\filtered\Negative\PermanentCrop_Neg"

os.makedirs(output_folder, exist_ok=True)

for img_path in Path(input_folder).glob("*.jpg"):
    img = cv2.imread(str(img_path), cv2.IMREAD_COLOR)
    if img is None:
        print(f"Skipping {img_path.name}")
        continue

    negative = 255 - img

    out_path = Path(output_folder) / (img_path.stem + "_negative.png")
    cv2.imwrite(str(out_path), negative)
    print(f"Done: {img_path.name}")

print("All done.")
```

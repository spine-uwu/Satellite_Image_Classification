# Dataset Preparation
Collects JPG images from two source folders, shuffles them, converts to PNG, and splits into **train / test / validation** at a **6 : 2 : 2** ratio.

## 1 · Install dependency

```python
!pip install Pillow -q
```

## 2 · Imports

```python
import os
import random
from pathlib import Path
from PIL import Image
```

## 3 · Configuration — edit these paths

```python
SRC_FOLDER_1 = "/path/to/folder1"   # <-- change this
SRC_FOLDER_2 = "/path/to/folder2"   # <-- change this
OUTPUT_BASE  = "/path/to/output"    # <-- change this

SPLIT_RATIO  = (6, 2, 2)            # train : test : validation
RANDOM_SEED  = 42
```

## 4 · Collect all JPG images

```python
exts = {".jpg", ".jpeg"}
all_images = []

for folder in [SRC_FOLDER_1, SRC_FOLDER_2]:
    p = Path(folder)
    if not p.is_dir():
        raise NotADirectoryError(f"Not a directory: {folder}")
    found = [f for f in p.iterdir() if f.suffix.lower() in exts]
    print(f"{folder}: {len(found)} JPG file(s) found")
    all_images.extend(found)

print(f"\nTotal images collected: {len(all_images)}")
```

## 5 · Shuffle

```python
random.seed(RANDOM_SEED)
random.shuffle(all_images)
print(f"Shuffled {len(all_images)} images with seed={RANDOM_SEED}")
```

## 6 · Split 6 : 2 : 2

```python
total  = sum(SPLIT_RATIO)
n      = len(all_images)
n_train = round(n * SPLIT_RATIO[0] / total)
n_test  = round(n * SPLIT_RATIO[1] / total)

train = all_images[:n_train]
test  = all_images[n_train : n_train + n_test]
val   = all_images[n_train + n_test :]          # remainder goes to validation

print(f"train      : {len(train)}")
print(f"test       : {len(test)}")
print(f"validation : {len(val)}")
print(f"total      : {len(train)+len(test)+len(val)}")
```

## 7 · Convert JPG → PNG and save

```python
splits = {
    "train":      train,
    "test":       test,
    "validation": val,
}

base         = Path(OUTPUT_BASE)
total_ok     = 0
total_fail   = 0
global_index = 0

for split_name, paths in splits.items():
    dest_dir = base / split_name
    dest_dir.mkdir(parents=True, exist_ok=True)
    ok = fail = 0

    for src in paths:
        dest = dest_dir / f"{global_index:05d}.png"
        try:
            with Image.open(src) as img:
                img.convert("RGB").save(dest, format="PNG")
            ok += 1
        except Exception as e:
            print(f"  [WARN] {src.name}: {e}")
            fail += 1
        global_index += 1

    print(f"[{split_name:>10}]  saved: {ok}  |  failed: {fail}  →  {dest_dir}")
    total_ok   += ok
    total_fail += fail

print(f"\n✓ Done — converted: {total_ok}  |  failed: {total_fail}")
print(f"Output base: {base.resolve()}")
```

## 8 · Quick sanity check

```python
for split_name in ["train", "test", "validation"]:
    folder = base / split_name
    count  = len(list(folder.glob("*.png")))
    print(f"{split_name:>10}: {count} PNG file(s)")
```
